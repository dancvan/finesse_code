```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
from pykat import finesse
from pykat.commands import *
from IPython.display import display, HTML

%matplotlib inline
pykat.init_pykat_plotting(dpi=90)
```

```python
#Amplitude modulation example
basekat =finesse.kat()
basekat.verbose = False

code = """

l laser 1 0 n0
s s1 1 n0 n1

mod eom 1k 0.3 1 am 0 n1 n2

s s2 1000 n2 n3 

"""
basekat.parseKatCode(code)
```

```python
kat1 = deepcopy(basekat)

code = """

ad car 0 n3
ad sb 1k n3

xaxis eom midx lin 0 1 200 #varying modulation index between 0 and 1
yaxis abs                  # mag of detector outputs

"""
kat1.parseCommands(code)
out1 = kat1.run()
```

```python
fig1 = out1.plot(title = 'Amplitude of freq.components vs mi', 
                 ylabel = 'Amplitude $[\sqrt{W}]$', 
                 xlabel = 'Modulation index')
```

```python
#can see that as the mi increases we see the amplitude of the carrier
#get fed into the sideband
```

```python
#Conservation of energy
#need to do the same thing as they did with the phase modulation 
kat2 = deepcopy(basekat)

code = """

ad car 0 n3
ad up 1k n3
ad low -1k n3

pd pow n3

xaxis eom midx lin 0 1 200
yaxis abs
"""

kat2.parseCommands(code)
out2 = kat2.run()
```

```python
#Power from the resulting fields
Pc =out2['car']**2
Ptot = (out2.y[:,0:3]**2).sum(axis=1)
Psb = (out2.y[:,1:3]**2).sum(axis=1)
```

```python
fig2 = plt.figure(figsize= (10,4))
ax = plt.subplot(1,1,1)
ax.plot(out2.x, Pc, label='Carrier')
ax.plot(out2.x, Ptot, label= 'Carrier and sidebands')
ax.plot(out2.x, Psb, label = 'Only sidebands')
ax.plot(out2.x, out2['pow'], 'k--', label ='Photo diode')

ax.set_title('Power conservation of EOM performing am')
ax.set_ylabel('Power [W]')
ax.set_xlabel('Modulation index')
ax.legend(loc=1)
plt.show(fig2)
```

```python
#notice that the DC part is getting reduced! 
# this is because the stronger the modulation, the more the larger
# part of the amplitude is fluctuating which leaves a smaller DC pow
# this leaves an overall apparent reduction of the power of the 3
# oscillations
```
