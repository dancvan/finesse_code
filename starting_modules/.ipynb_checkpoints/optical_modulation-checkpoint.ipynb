```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
from pykat import finesse
from pykat.commands import * 
from IPython.display import display, HTML 

%matplotlib inline

pykat.init_pykat_plotting(dpi=90)
```

```python
#phase modulation

t = np.linspace(0,1,2048)
E0 = 1 #electric field 
fc = 20 #carrier freq
phi_c = 0 #phase of the field
fm = 2 #phase modulation
m = 5 #modulation index
phi_m = 0 #modulation phase

E= E0*np.cos(fc*2*np.pi*t + phi_c) #carrier field

x = m*np.sin(fm*2*np.pi*t + phi_m) # signal

E_m = E0*np.cos(fc*2*np.pi*t + phi_c + m*np.sin(fm*2*np.pi*t + phi_m))

#plot the shit
fig = plt.figure(figsize=(10,3))

ax = plt.subplot(1,1,1)
p1 = ax.plot(t,E,'0.7', label='$\mathrm{Carrier}\ E_c(t)$')
p2 = ax.plot(t,E_m,'r', label='$\mathrm{Mod. field}\ E_m(t)$')

ax.set_ylabel('E(t)')
ax.set_xlabel('t')
ax.set_title('Phase modulation')
ax.set_xlim(-0.01, 1.01)

ax2 = ax.twinx()
p3 = ax2.plot(t,x,'b', label='$\mathrm{Signal}\ x(t)$')
ax2.set_ylabel('x(t)')

plots = p1 + p2 + p3 
labs = [lab.get_label() for lab in plots]
ax.legend(plots, labs, loc=1, fontsize=10)

plt.show(fig)
```

```python
#phase moulation in practice
basekat = finesse.kat()
basekat.verbose = False

code = """

l laser 1 0 n0 # laser with 0 offset wavelength
s s1 1 n0 n1   # spaaaaaace
mod eom 1k 0.3 5 pm 0 n1 n2 # 1kHz  phase modulation with mi of .3 
                            # 5 sideband pairs, and 0 degree modulation
                            # phase
                            
s s2 1000 n2 n3 # more spaaaace (1m)
"""

basekat.parseKatCode(code)
```

```python
kat1 = basekat.deepcopy()

code = """

ad car 0 n3     # amplitude detector for the carrier frequency 
ad up1 1k n3    # amplitude ddetector for the first uppser sideband
ad up2 2k n3    # ad for the second upper sideband
ad up3 3k n3    # ....
ad up4 4k n3
ad up5 5k n3
#ad up6 6k n3

xaxis eom midx lin 0 6 200 #varying mi between 0 and 1
yaxis abs        # showing amplitudes

"""

kat1.parseCommands(code)
out1 = kat1.run()
```

```python
fig1 = out1.plot(title = 'Amplitude of the frequency components vs mi',
                 ylabel='Amplitude $[\sqrt{W}$]',
                 xlabel = 'Modulation index')
```

```python
# This particular distribution of amplitude is characterized 
# mathematically by a taylor expansion about m=0 using a truncated
# sum of Bessel-functions
```

```python
# Conservation of Energy

# the idea I want to execute here is to add up all of the power 
# that is distributed into the sidebands

kat2 = basekat.deepcopy()

code = """


ad car 0 n3     #carrier
ad up1 1k n3    # first upper sideband
ad low1 -1k n3   # first lower sideband 

ad up2 2k n3    # second upper sideband
ad low2 -2k n3   # second lower sideband

ad up3 3k n3    # third upper sideband
ad low3 -3k n3   # third lower sideband

ad up4 4k n3    # fourth upper sideband
ad low4 -4k n3   # fourth lower sideband

ad up5 5k n3    # fifth upper sideband
ad low5 -5k n3   # fifth lower sideband

#ad up6 6k n3    # sixth upper sideband
#ad low6 -6k n3   # sixth lower sideband

pd pow n3        # DC power readout

xaxis eom midx lin 0 6 200 # varying mi between 0 and 6
yaxis abs         # absolute value of the detector outputs

"""
kat2.parseKatCode(code)
out2 = kat2.run()
```

```python
print(out2.ylabels)
print(out2.y.shape)
```

```python
#Power calc
P1 = (out2.y[:,0:3]**2).sum(axis=1)
P2 = (out2.y[:,0:5]**2).sum(axis=1)
P3 = (out2.y[:,0:7]**2).sum(axis=1)
P4 = (out2.y[:,0:9]**2).sum(axis=1)
P5 = (out2.y[:,0:11]**2).sum(axis=1)
#P6 = (out2.y[:,0:13]**2).sum(axis=1)

fig2 = plt.figure(figsize=(8,4))
ax = plt.subplot(1,1,1)
ax.plot(out2.x, P1, label = '1 sb pairs')
ax.plot(out2.x, P2, label = '2 sb pairs')
ax.plot(out2.x, P3, label = '3 sb pairs')
ax.plot(out2.x, P4, label = '4 sb pairs')
ax.plot(out2.x, P5, label = '5 sb pairs')
#ax.plot(out2.x, P6, label = '6 sb pairs')
ax.plot(out2.x, out2['pow'], 'k--', label = 'DC Power')

ax.set_title('Power conservation of EOM')
ax.set_ylabel('Power [W]')
ax.set_xlabel('Modulation index')
ax.legend(loc=3)
plt.show(fig2)
```
