```python
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt
from pykat import finesse
from pykat.commands import * 
from IPython.display import HTML, display

%matplotlib inline

pykat.init_pykat_plotting(dpi = 90)

```

```python
#create your finesse code
basekat = finesse.kat()
basekat.verbose = False
basecode = """
l laser 1 0 n0  # laser (1 W, no wavelength offset (0))
s s1 1 n0 nc1     #free space 1 meter (n0 and nc1: name of two nodes)


# the cavity
m m1 0.7 0.3 0 nc1 nc2 # (nc1 nc2 front and back of mirror)
s sL 4000 nc2 nc3 # Space (length of cavity)
m m2 0.8 0.2 0 nc3 nc4
"""
basekat.parseKatCode(basecode) #parsing the FINESSE-code
```

```python
kat1 = deepcopy(basekat)
code= """
#adding detectors


#DC pds
pd refl nc1 # cavity refl
pd trans nc4 # cavity tranmission
pd circ nc3  #circulated power at node 1
pd circ_2 nc2 #circulated power at node 2

#simulation instructions
xaxis m1 phi lin 0 450 2000 #Vary the tuning of m1
yaxis abs
"""

kat1.parseKatCode(code)
out1 = kat1.run()

```

```python

fig_1 = out1.plot(xlabel = 'Position of input mirror [deg]',
                  ylabel = 'Power [W]', 
                  title = "Power vs microscopic cavity tuning")
```

```python
#Seeing the power change over macroscopic lengths
kat2 = deepcopy(basekat)
code= """
#adding detectors


#DC pds
pd refl nc1 # cavity refl
pd trans nc4 # cavity tranmission
pd circ nc3  #circulated power at node 1
pd circ_2 nc2 #circulated power at node 2

#simulation instructions
xaxis sL L lin 0 10k 10000 #Vary the tuning of m1
yaxis abs
"""

kat2.parseKatCode(code)
out2 = kat2.run()
```

```python
fig_2 = out2.plot(xlabel = 'increasing length between cavity mirrors[m]',
                  ylabel = 'Power [W]', 
                  title = "Power vs microscopic cavity tuning")
```

```python
# As you can see... finessse still thinks we're constantly on resonance
# this may be a bit problematic with intuition since you think that 
#realistically, through macroscopic length changes you would sweep through
#resonances... but you have to differentiate between macropscopic and
#microscopic length changes
```

```python
#What happens if we modify the laser frequency from it's 0 offset?
kat3 = deepcopy(basekat)
code= """
#adding detectors


#DC pds
pd refl nc1 # cavity refl
pd trans nc4 # cavity tranmission
pd circ nc3  #circulated power at node 1
pd circ_2 nc2 #circulated power at node 2

#simulation instructions
xaxis laser f lin 0 100k 1000 #Vary the tuning of m1
yaxis abs
"""

kat3.parseKatCode(code)
out3 = kat3.run()

```

```python
fig_3 = out3.plot(xlabel = 'Sweeping the laser frequency going into the cavity [Hz]',
                  ylabel = 'Power [W]', 
                  title = "Power vs microscopic cavity tuning")
```

```python
# as you see, it is equivalent to having a microscopic change in length
```

```python
#Changing the frequency offset to 20kHz an then make a macroscopic change in length
kat4 = kat2.deepcopy()
kat4.laser.f = 20e3
out4 = kat4.run()
fig_4 = out4.plot(xlabel = 'cavity length [m]',
                  ylabel = 'Power [W]', 
                  title = "Power vs macroscopic cavity tuning (w/ 20kHz laser freq offset)")
```

```python
# The idea here is that we saw that sweeping the frequency essentially 
#caused the same effect as the microscopic detuning... with a little bit 
#of frequency offset, we are likely to see macroscopic resonances due to
#small frequency offset (relative to the carrier frequency)
```
