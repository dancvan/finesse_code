```python
import numpy as np
import matplotlib  
import matplotlib.pyplot as plt
from pykat import finesse
from pykat.commands import *
from IPython.display import display, HTML

%matplotlib inline

pykat.init_pykat_plotting(dpi=90)
```

```python
basekat = finesse.kat()
basekat.verbose = False
basecode = """

l laser 1 0 n0
s s1 1 n0 n1 

mod eom 1k .3 1 am 30 n1 n2  #Phase this time is 30 degrees


s s2 1n n2 n3   # space EOM to output
"""
basekat.parseCommands(basecode)
```

```python
#Will be placing down 3 photodiodes at the same exact node 
# but we will be measuring at three different demodulation phases
# phi_x = 0 (in-phase) 
# phi_x = 90 (in-quadrature)
# phi_x = 30 (modulation phase we use)
#after demodulation (or mixing) we low pass our signals so that the 
#DC parts are the only ones that remain
#The photodiode and the mixing and the low passing are all done with 
#pd1 where 1 stands for single demodulation
# while doing this we are varying the modulation index of the EOM
```

```python
kat1 = basekat.deepcopy()

code = """

pd1 inphase 1k 0 n3  #pd + mixer + lpf (demod 1kHz demod phase 0 deg)

pd1 inquad 1k 90 n3 

pd1 optimal 1k 30 n3

xaxis eom midx lin 0 1 200
yaxis abs
"""

kat1.parseCommands(code)
out1 = kat1.run()

fig1a = out1.plot(xlabel = 'MI', ylabel = 'Powaaaah [W]')
```

```python
# the optimal one assumes that we know the modulation phase to begin
# with.. this is true in our case but realistically this is not the 
# case.... Let's reconstruct the full demodulation signal
```

```python
tot = out1['inphase'] + 1j*out1['inquad'] # S_1 = S_(1,phase) + S_(1,quad)
fig1b = plt.figure(figsize = (8,4))
ax = plt.subplot(1,1,1)
ax.plot(out1.x, out1['inphase'], 'b-', label = 'In-phase (0 deg)')
ax.plot(out1.x, out1['inquad'], 'k-', label = 'In-quadrature (90 deg)')
ax.plot(out1.x, out1['optimal'], 'r-',label = 'Optimal')
ax.plot(out1.x, np.abs(tot), 'k--', label = 'Reconstructed')

ax.set_xlabel('Modulation index $m$')
ax.set_ylabel('Power [W]')
ax.set_title('Demodulation')
ax.legend(loc=0, fontsize=10)
plt.show()
```

```python
#LOOK MA! I reconstructed the signal!
```

```python
#Reconstructing the COMPLEX AMPLITUDE OF THE SIDEBAND

#want to pretty much copy the code above except we want to add a 
# dc photodiode to measure the DC power and an amplitude detector to 
# measure the complex amplitude of the upper sidband. 

kat2 = deepcopy(basekat)

code = """
pd1 inphase 1k 0 n3   # pd + mixer + lpf 
pd1 inquad 1k 90 n3   #pd + mixer + lpf (demod phase 90 degrees)

pd dc n3          #pd + lpf

ad amp 1k n3     # amplitude detector

xaxis eom midx lin 0 1 400 #varying eom mi between 0 and 1
yaxis abs:deg              # returns absolute values of detector outputs
"""
kat2.parseCommands(code)
out2 = kat2.run()
```

```python
# the DC signal is given by S_(DC) = |a_o|^2 + |a_-|^2 + |a_+|^2
# Full complex demodulation signal is given by S_1 = (a_o)(a*_-) + (a*_0)(a_+)
# IN our case the carrier has zero frequency (as long as it is not
# transmitted through any media)
# Therefore the DC power can be written as S_(DC) = |a_o|^2 + 2|a_+|^2
# Demodulation signal is S_1 = 2(a_0)(a_+)
#Combining the two we can calculate for a_o  and a_+
# We can now use Finesse to calculate the sideband amplitude
```

```python
S1 = out2['inphase'] + 1j*out2['inquad']

a0 = np.sqrt( (out2['dc'] + np.sqrt(out2['dc']**2 - 2*np.abs(S1)**2))/2)

a_plus = S1/(2*a0)


fig1b = plt.figure(figsize =(8,6))
ax = plt.subplot(2,1,1)
ax.plot(out2.x, np.abs(out2['amp']), 'r-', label = 'Amplitude')
ax.plot(out2.x, np.abs(a_plus), 'k--', label = 'Reconstructed')
ax.set_ylabel('Amplitude[$\sqrt{\mathrm{W}}$]')
ax.set_title('Complex amplitude of upper sideband')
ax.legend(loc=0, fontsize=10)

ax2 =plt.subplot(2,1,2)
ax2.plot(out2.x, np.angle(out2['amp'])*180/np.pi, 'r-', label='Amplitude')
ax2.plot(out2.x, np.angle(a_plus)*180/np.pi, 'k--', label = 'Reconstructed')
ax2.set_xlabel('Modulation index $m$')
ax2.set_ylabel('Phase [deg]')
ax2.set_ylim((0,35))
ax2.legend(loc=0, fontsize=10)

plt.show()
```
